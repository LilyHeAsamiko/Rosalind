```python
def isInterleave(s1: str, s2: str, s3: str) -> bool:
    """s1与s2能否交织构成s3"""
    n, m1, m2 = len(s3), len(s1), len(s2)
    if m1 + m2 != n:
        return False
    dp = [[False] * (m2 + 1) for _ in range(m1 + 1)]  # 是长度，不是索引
    # 状态初始化
    dp[0][0] = True  # 两者长度都为 0 时为 True
    for j in range(1, m2 + 1):  # len(s1) = 0，枚举 s2 的长度
        dp[0][j] = dp[0][j-1] and s3[j-1] == s2[j-1]  # 注意索引
    for i in range(1, m1 + 1):  # len(s2) = 0，枚举 s1 的长度
        dp[i][0] = dp[i-1][0] and s3[i-1] == s1[i-1]  # 注意索引
    # 状态转移：直接从长度 >= 1 开始转移
    for i in range(1, m1 + 1):
        for j in range(1, m2 + 1):
            p = i + j - 1  # s3 对应字符的索引
            dp[i][j] = (s3[p] == s1[i-1] and dp[i-1][j]) or (s3[p] == s2[j-1] and dp[i][j-1])
    return dp[m1][m2]
```

```python
def disjointMotifs(DNA: str, motifs: List[str]) -> int:
    """枚举motif两两组合，检测它们是否为DNA某一等长子串的交织序列"""
    n = len(motifs)
    ans = [['0'] * n for _ in range(n)]
    from collections import Counter
    # 枚举两两序列对
    for i in range(n):
        for j in range(i, n):
            m = len(motifs[i]) + len(motifs[j])
            # 枚举 DNA 长度为 m 的子串 s 是否由 motifs[i]和motifs[j] 交织而成
            for k in range(len(DNA) - m + 1):
                if isInterleave(motifs[i], motifs[j], DNA[k:k+m]):
                    ans[i][j] = ans[j][i] = '1'
    return ans
```

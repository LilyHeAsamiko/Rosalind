```python
import re
def create_a_character_table(newick):

    taxa = sorted(taxon for taxon in re.split('[(),]+', newick.rstrip(';')) if taxon)

    # find all the locations of pairs of matching parentheses
    stack = []
    match = []
    for j in range(len(newick)):
        if newick[j] == '(':
            stack.append(j)
        if newick[j] == ')':
            i = stack.pop()
            match.append((i, j))

    table = []
    for (i, j) in match[:-1]:
        subset = [taxon for taxon in re.split('[(),]+', newick[i:j+1]) if taxon]
        array = [int(taxa[k] in subset) for k in range(len(taxa))]
        table.append(''.join(map(str, array)))
    return table
```

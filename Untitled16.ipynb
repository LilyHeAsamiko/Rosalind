```python
def distance_check(T):
  ct = ''
  for i in T:
        ct += i
  while '(,)' in ct:
        ct = ct.replace('(,)','')
  if ct.count('(') == len(ct):
        return len(ct)
  elif ct.count(')') == len(ct):
        return len(ct)
  elif ct.count(',') == len(ct):
        return 2
  else:
        return ct.count(')') + ct.count('(') + 2
```
